## cross-match con gaia_edr3 por medio de coordenadas

```python
query = f"""
      SELECT source_id, gaia.ra, gaia.dec
      FROM user_{user_name}.{table_name} AS chandra
      JOIN gaiaedr3.gaia_source AS gaia
      -- Geometric Cross-Match: =======
      ON 1 = CONTAINS(
      POINT(chandra.ra, chandra.dec),
      CIRCLE(gaia.ra, gaia.dec, 4/3600.))"""
```

```python
job= Gaia.launch_job_async(query)
results_match = job.get_results()

print(f'Table size (rows): {len(results_match)}')
match = results_match.to_pandas()
```

## Clasificación no supersivada GaussianMixture para datos de gaia

Como gaia registras más fuentes, entonces puede que el algortimo de GM funcione mejor.

```python
n=9 #número de grupos
gmm = GMM(n, covariance_type='full', random_state=0)

gaia_results.dropna(inplace=True)


gmm.fit(gaia_results[["ra", "dec"]])
gaia_results['label']= gmm.predict(gaia_results[["ra", "dec"]])


cs = [gaia_results[gaia_results['label']==i] for i in range(n)]
centros= [np.array([c['ra'].mean(), c['dec'].mean()]) for c in cs]
```

```python
plt.figure(figsize=(10,10))
plt.title("NGC5236")

for i,c in enumerate(cs):
    plt.plot(c['ra'], c['dec'], ".", label=f'{i}')
    plt.plot(centros[i][0], centros[i][1], "*k", ms=10)


#plt.plot(m83.ra.value, m83.dec.value, "b*")
plt.axis('equal')
plt.legend()
plt.show()
```

## Definición de los brazos

```python
up_arm= [gaia_results.query(f'label=={i}') for i in [1,5,6,3]]

down_arm= [gaia_results.query(f'label=={i}') for i in [4,8,0]]
```

## interpolación up_arm

```python
centros_up= [np.array([c['ra'].mean(), c['dec'].mean()]) for c in up_arm]

centros_down= [np.array([c['ra'].mean(), c['dec'].mean()]) for c in down_arm]
```

## calculo coordenadas polares de los centros de cada grupo

coordenadas respecto al centro de la galaxia

```python
flag_r= False #interpolación de theta en función de r
```

```python
ccp_
```

```python
centros_c_up= [c - np.array([m83.ra.value, m83.dec.value]) for c in centros_up]
centros_c_down= [c - np.array([m83.ra.value, m83.dec.value]) for c in centros_down]


from math import atan2

centros_c_up_r= [np.sqrt(c[0]**2 + c[1]**2) for c in centros_c_up]
centros_c_down_r= [np.sqrt(c[0]**2 + c[1]**2) for c in centros_c_down]


centros_c_up_theta= [ atan2(c[1], c[0]) for c in centros_c_up]
centros_c_up_theta[-1]+= 2*np.pi
centros_c_down_theta= [ atan2(c[1], c[0]) for c in centros_c_down]

if flag_r:
    
    centros_c_up_polar = np.array(list(zip(centros_c_up_r, centros_c_up_theta)))
    centros_c_down_polar = np.array(list(zip(centros_c_down_r, centros_c_down_theta)))
else:
    centros_c_up_polar = np.array(list(zip(centros_c_up_theta, centros_c_up_r)))
    centros_c_down_polar = np.array(list(zip(centros_c_down_theta, centros_c_down_r)))
    
if flag_r:

    ccp_up = pd.DataFrame(centros_c_up_polar, columns= ["r", "theta"] )
    ccp_up.sort_values('r', inplace=True)
    
    ccp_down = pd.DataFrame(centros_c_down_polar, columns= ["r", "theta"] )
    ccp_down.sort_values('r', inplace=True)

else:

    ccp_up = pd.DataFrame(centros_c_up_polar, columns= ["theta", "r"] )
    ccp_up.sort_values('theta', inplace=True)
    
    ccp_down = pd.DataFrame(centros_c_down_polar, columns= ["theta", "r"] )
    ccp_down.sort_values('theta', inplace=True)
```

## Interpolación de los centros de los grupos

```python
from scipy.interpolate import CubicSpline
from scipy.interpolate import interp1d
```

```python
if flag_r:
    inter_cs_up= CubicSpline(ccp_up['r'], ccp_up['theta'])
    inter_1d_up= interp1d(ccp_up['r'], ccp_up['theta'], kind='linear')
    
    inter_cs_down= CubicSpline(ccp_down['r'], ccp_down['theta'])
    inter_1d_down= interp1d(ccp_down['r'], ccp_down['theta'], kind='linear')
else:
    inter_cs_up= CubicSpline(ccp_up['theta'], ccp_up['r'])
    inter_1d_up= interp1d(x= ccp_up['theta'], y= ccp_up['r'],  kind='linear')
    
    inter_cs_down= CubicSpline(ccp_down['theta'], ccp_down['r'])
    inter_1d_down= interp1d(x= ccp_down['theta'], y= ccp_down['r'],  kind='linear')
    
    
m=10000 #particiones

if flag_r:

    rs_inter_up= np.linspace(np.min(centros_c_up_r),np.max(centros_c_up_r), m)
    rs_inter_down= np.linspace(np.min(centros_c_down_r),np.max(centros_c_down_r), m)

else:

    thetas_inter_up= np.linspace(np.min(ccp_up['theta']), np.max(ccp_up['theta']), m)
    thetas_inter_down= np.linspace(np.min(ccp_down['theta']), np.max(ccp_down['theta']), m)
    
if flag_r:
    thetas_cs_up= inter_cs_up(rs_inter_up)
    thetas_1d_up= inter_1d_up(rs_inter_up)
    
    thetas_cs_down= inter_cs_down(rs_inter_down)
    thetas_1d_down= inter_1d_down(rs_inter_down)
else:
    rs_cs_up= inter_cs_up(thetas_inter_up)
    rs_1d_up= inter_1d_up(thetas_inter_up)
    
    rs_cs_down= inter_cs_down(thetas_inter_down)
    rs_1d_down= inter_1d_down(thetas_inter_down)
    

if flag_r:
    plt.plot(rs_inter_up,thetas_cs_up, ".-", label= 'cs')
    plt.plot(rs_inter_up,thetas_1d_up, ".-", label="1d")
    plt.plot(ccp_up['r'],ccp_up['theta'],".")
    
    plt.plot(rs_inter_down,thetas_cs_down, ".-", label= 'cs')
    plt.plot(rs_inter_down,thetas_1d_down, ".-", label="1d")
    plt.plot(ccp_down['r'],ccp_down['theta'],".")
    

else:
    
    plt.plot(thetas_inter_up,rs_cs_up, ".-", label= 'cs')
    plt.plot(thetas_inter_up,rs_1d_up, ".-", label= '1d')
    plt.plot(ccp_up['theta'],ccp_up['r'],".")
    

    plt.plot(thetas_inter_down,rs_cs_down, ".-", label= 'cs')
    plt.plot(thetas_inter_down,rs_1d_down, ".-", label= '1d')
    plt.plot(ccp_down['theta'],ccp_down['r'],".")


plt.legend()
plt.show()


if flag_r:
    ras_1d_up= rs_inter_up*np.cos(thetas_1d_up) + m83.ra.value
    decs_1d_up= rs_inter_up*np.sin(thetas_1d_up) + m83.dec.value
    
    ras_1d_down= rs_inter_down*np.cos(thetas_1d_down) + m83.ra.value
    decs_1d_down= rs_inter_down*np.sin(thetas_1d_down) + m83.dec.value

    ras_cs_up= rs_inter_up*np.cos(thetas_cs_up) + m83.ra.value
    decs_cs_up= rs_inter_up*np.sin(thetas_cs_up) + m83.dec.value
    
    ras_cs_down= rs_inter_down*np.cos(thetas_cs_down) + m83.ra.value
    decs_cs_down= rs_inter_down*np.sin(thetas_cs_down) + m83.dec.value

else:
  
    ras_1d_up= rs_1d_up*np.cos(thetas_inter_up) + m83.ra.value
    decs_1d_up= rs_1d_up*np.sin(thetas_inter_up) + m83.dec.value
    
    ras_1d_down= rs_1d_down*np.cos(thetas_inter_down) + m83.ra.value
    decs_1d_down= rs_1d_down*np.sin(thetas_inter_down) + m83.dec.value

    ras_cs_up= rs_cs_up*np.cos(thetas_inter_up) + m83.ra.value
    decs_cs_up= rs_cs_up*np.sin(thetas_inter_up) + m83.dec.value
    
    ras_cs_down= rs_cs_down*np.cos(thetas_inter_down) + m83.ra.value
    decs_cs_down= rs_cs_down*np.sin(thetas_inter_down) + m83.dec.value
```

## Transfomación a coordenadas polares de los centros

```python
centros_c= [c - np.array([m83.ra.value, m83.dec.value]) for c in centros]

flag_r= True #interpolación de theta en función de r

from math import atan2

centros_c_r= [np.sqrt(c[0]**2 + c[1]**2) for c in centros_c]
centros_c_theta= [ atan2(c[1], c[0]) for c in centros_c]

if flag_r:
    centros_c_polar = np.array(list(zip(centros_c_r, centros_c_theta)))
else:
    centros_c_polar = np.array(list(zip(centros_c_theta, centros_c_r)))
    
if flag_r:

    ccp = pd.DataFrame(centros_c_polar, columns= ["r", "theta"] )
    ccp.sort_values('r', inplace=True)

else:

    ccp = pd.DataFrame(centros_c_polar, columns= ["theta", "r"] )
    ccp.sort_values('theta', inplace=True)

ccp.reset_index(inplace=True)


ccp.head()
```

## Interpolación

```python
if flag_r:
    inter_cs= CubicSpline(ccp['r'], ccp['theta'])
    inter_1d= interp1d(ccp['r'], ccp['theta'], kind='linear')
else:
    inter_cs= CubicSpline(ccp['theta'], ccp['r'])
    inter_1d= interp1d(x= ccp['theta'], y= ccp['r'],  kind='linear')
    
m=10000 #particiones

if flag_r:

    rs_inter= np.linspace(np.min(centros_c_r),np.max(centros_c_r), m)

else:

    thetas_inter= np.linspace(np.min(ccp['theta']), np.max(ccp['theta']), m)
    
if flag_r:
    thetas_cs= inter_cs(rs_inter)
    thetas_1d= inter_1d(rs_inter)
else:
    rs_cs= inter_cs(thetas_inter)
    rs_1d= inter_1d(thetas_inter)
    

if flag_r:
    plt.plot(rs_inter,thetas_cs, ".-", label= 'cs')
    plt.plot(rs_inter,thetas_1d, ".-", label="1d")
    plt.plot(ccp['r'],ccp['theta'],".")

else:

    plt.plot(thetas_inter,rs_cs, ".-", label= 'cs')
    plt.plot(thetas_inter,rs_1d, ".-", label= '1d')
    plt.plot(ccp['theta'],ccp['r'],".")



plt.legend()
plt.show()


if flag_r:
    ras_1d= rs_inter*np.cos(thetas_1d) + m83.ra.value
    decs_1d= rs_inter*np.sin(thetas_1d) + m83.dec.value

    ras_cs= rs_inter*np.cos(thetas_cs) + m83.ra.value
    decs_cs= rs_inter*np.sin(thetas_cs) + m83.dec.value

else:
  
    ras_1d= rs_1d*np.cos(thetas_inter) + m83.ra.value
    decs_1d= rs_1d*np.sin(thetas_inter) + m83.dec.value

    ras_cs= rs_cs*np.cos(thetas_inter) + m83.ra.value
    decs_cs= rs_cs*np.sin(thetas_inter) + m83.dec.value
```
