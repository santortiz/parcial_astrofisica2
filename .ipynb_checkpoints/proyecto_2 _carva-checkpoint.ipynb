````python
def parse_stc_s(stc_s):
    """
    funcion para convertir el string con el que se
    muestran los polígonos a un arreglo de coordenadas.
    """
    sfov = stc_s.split(' ')[2:]
    ras = []
    decs = []
    for i, c in enumerate(sfov):
        if i%2==0:
            ras.append(c)
        else:
            decs.append(c)
    ras = np.array(ras, dtype=float)
    decs = np.array(decs, dtype=float)
    coords = np.array([ras,decs]).T
    return coords
````

### Nota importante:

Es evidente que estos conjuntos de datos no pueden ser representados por funciones, porque los mapas adecuados para esos ajustes no serían biyectivos en ningún eje, necesariamente.

Por esta razón, antes de realizar los ajustes, realizaré una transformación de coordenadas en las que las curvas sí puedan ser representadas con funciones (no paramétricas). Al tener funciones no-paramétricas, además de realizar una interpolación, podré realizar un ajuste polinomial, por ejemplo.

````python
# Parametrizando un ángulo y un r respecto al centro de la galaxia
# en el plano visual.

cra, cdec = C_NGC5236.ra.value, C_NGC5236.dec.value
def rade2rth(ra,dec,mod=True):
    x = ra-cra
    y = dec-cdec
    
    r = (x**2+y**2)**0.5
    
    # uso np.mod para obtener el resultado en el rango (0,2*pi)
    if mod:
        th = np.mod(np.arctan2(y*np.pi/180,x*np.pi/180),2*np.pi)
    else:
        th = np.arctan2(y*np.pi/180,x*np.pi/180)
    return r, th

def rth2rade(r,th):
    x = r*np.cos(th)
    y = r*np.sin(th)
    
    ra = x + cra
    dec = y+ cdec
    return ra, dec
````

````python
# calculando para cada dataset las transformaciones correspondientes.
brazo1['r'] = brazo1[['mra','mdec']].apply(lambda x: rade2rth(x[0],x[1],False)[0],axis=1)
brazo1['th'] = brazo1[['mra','mdec']].apply(lambda x: rade2rth(x[0],x[1],False)[1],axis=1)

brazo2['r'] = brazo2[['mra','mdec']].apply(lambda x: rade2rth(x[0],x[1])[0],axis=1)
brazo2['th'] = brazo2[['mra','mdec']].apply(lambda x: rade2rth(x[0],x[1])[1],axis=1)

brazo2f['r'] = brazo2f[['ra','dec']].apply(lambda x: rade2rth(x[0],x[1])[0],axis=1)
brazo2f['th'] = brazo2f[['ra','dec']].apply(lambda x: rade2rth(x[0],x[1])[1],axis=1)

brazo1f['r'] = brazo1f[['ra','dec']].apply(lambda x: rade2rth(x[0],x[1],False)[0],axis=1)
brazo1f['th'] = brazo1f[['ra','dec']].apply(lambda x: rade2rth(x[0],x[1],False)[1],axis=1)
````

````python
brazo1
````

````python
# organizando para poder hacer la interpolación por splines
brazo1.sort_values('th', inplace=True)
brazo2.sort_values('th', inplace=True)
````

````python
interp = 'poly3' # puede ser 'poly3' o 'cs'

if interp=='cs':
    thr1 = np.linspace(min(brazo1['th']),max(brazo1['th']),200)
    thr2 = np.linspace(min(brazo2['th']),max(brazo2['th']),200)
    
    cs1 = CubicSpline(brazo1['th'], brazo1['r'])
    rr1 = cs1(thr1)

    cs2 = CubicSpline(brazo2['th'], brazo2['r'])
    rr2 = cs2(thr2)
    
if interp=='poly3':
    thr1 = np.linspace(min(brazo1f['th']),max(brazo1['th']),200)
    thr2 = np.linspace(min(brazo2f['th']),max(brazo2f['th']),200)
    
    z1 = np.polyfit(brazo1f['th'],brazo1f['r'],3)
    poly1 = np.poly1d(z1)
    rr1 = poly1(thr1)
    
    z2 = np.polyfit(brazo2f['th'],brazo2f['r'],3)
    poly2 = np.poly1d(z2)
    rr2 = poly2(thr2)
````

## Definición de los brazos

````python
up_arm= [gaia_results.query(f'label=={i}') for i in [1,5,6,3]]

down_arm= [gaia_results.query(f'label=={i}') for i in [4,8,0]]
````

````python
centros_up= [np.array([c['ra'].mean(), c['dec'].mean()]) for c in up_arm]

centros_down= [np.array([c['ra'].mean(), c['dec'].mean()]) for c in down_arm]
````

## Paso a coordenadas polares

````python
centros_c_up= [c - np.array([m83.ra.value, m83.dec.value]) for c in centros_up]
centros_c_down= [c - np.array([m83.ra.value, m83.dec.value]) for c in centros_down]

centros_c_up_r= [np.sqrt(c[0]**2 + c[1]**2) for c in centros_c_up]
centros_c_down_r= [np.sqrt(c[0]**2 + c[1]**2) for c in centros_c_down]


centros_c_up_theta= [ atan2(c[1], c[0]) for c in centros_c_up]
centros_c_up_theta[-1]+= 2*np.pi
centros_c_down_theta= [ atan2(c[1], c[0]) for c in centros_c_down]

centros_c_up_polar = np.array(list(zip(centros_c_up_theta, centros_c_up_r)))
centros_c_down_polar = np.array(list(zip(centros_c_down_theta, centros_c_down_r)))

ccp_up = pd.DataFrame(centros_c_up_polar, columns= ["theta", "r"] )
ccp_up.sort_values('theta', inplace=True)
    
ccp_down = pd.DataFrame(centros_c_down_polar, columns= ["theta", "r"] )
ccp_down.sort_values('theta', inplace=True)
````

## Interpolación de r en función de theta

````python
inter_cs_up= CubicSpline(ccp_up['theta'], ccp_up['r'])
inter_1d_up= interp1d(x= ccp_up['theta'], y= ccp_up['r'],  kind='linear')
    
inter_cs_down= CubicSpline(ccp_down['theta'], ccp_down['r'])
inter_1d_down= interp1d(x= ccp_down['theta'], y= ccp_down['r'],  kind='linear')

m=10000 #particiones

thetas_inter_up= np.linspace(np.min(ccp_up['theta']), np.max(ccp_up['theta']), m)
thetas_inter_down= np.linspace(np.min(ccp_down['theta']), np.max(ccp_down['theta']), m)

rs_cs_up= inter_cs_up(thetas_inter_up)
rs_1d_up= inter_1d_up(thetas_inter_up)

rs_cs_down= inter_cs_down(thetas_inter_down)
rs_1d_down= inter_1d_down(thetas_inter_down)

plt.figure()
plt.title("Curvas de interpolación")

plt.plot(thetas_inter_up,rs_cs_up, ".-", label= 'cs')
plt.plot(thetas_inter_up,rs_1d_up, ".-", label= '1d')
plt.plot(ccp_up['theta'],ccp_up['r'],".")


plt.plot(thetas_inter_down,rs_cs_down, ".-", label= 'cs')
plt.plot(thetas_inter_down,rs_1d_down, ".-", label= '1d')
plt.plot(ccp_down['theta'],ccp_down['r'],".")

plt.legend()
plt.show()
````
